#### 🖥️ Vista General (Overview)

**S (Situación):**  
Andes Retail Group registró aproximadamente 6 millones en ingresos entre 2024 y 2025, con 1.94 millones en ganancia, 58 mil unidades vendidas y alrededor de 4 mil clientes únicos. Las ventas están distribuidas principalmente entre Perú, Chile y Colombia, con una participación importante de los segmentos Premium y Estándar.

**C (Complicación):**  
Aunque los ingresos totales se mantienen sólidos, el desempeño no es uniforme:

    - Existe dependencia significativa de los segmentos Premium y Estándar.
    - El segmento Económico contribuye solo 0.5M.
    - Además, la evolución mensual muestra variaciones importantes, con varios meses por debajo del promedio.

Esto sugiere volatilidad en la demanda y dependencia de ciertos mercados y segmentos.


**Q (Pregunta):**   
¿Cómo puede la empresa mantener el crecimiento de ingresos y mejorar la estabilidad de ventas, reduciendo la dependencia de segmentos específicos y optimizando los mercados más rentables?


**A (Respuesta):**  
Los datos muestran que:

    - Perú y Chile son los principales motores de ingresos
    - El segmento Premium es el mayor generador de valor
    - Existen meses con desempeño inferior al promedio que representan oportunidades de optimización comercial.

Por lo tanto, se recomienda:

    - Priorizar estrategias comerciales en segmentos Premium
    - Expandir el desempeño del segmento Estándar
    - Implementar campañas para fortalecer ventas en meses de baja demanda

#### 🔎 Vista Detalle

**S (Situación):**  
El análisis detallado permite evaluar el desempeño de ventas por categoría de producto, región geográfica y estacionalidad, proporcionando visibilidad sobre qué áreas impulsan el crecimiento del negocio.

Las categorías presentan una distribución relativamente equilibrada entre Deportes, Electrónica, Hogar y Ropa, mientras que las regiones Norte, Centro y Sur muestran contribuciones similares.

**C (Complicación):**  
A pesar de la distribución equilibrada entre categorías y regiones, el comportamiento estacional revela una fuerte dependencia del verano, que genera aproximadamente 2.2 millones en ingresos, mientras que el invierno cae a 0.7 millones.

Esto indica alta estacionalidad del negocio, lo cual puede generar fluctuaciones significativas en los resultados anuales.

**Q (Pregunta):**  
¿Qué oportunidades existen para reducir la estacionalidad de ventas y optimizar el rendimiento por categoría, región y temporada?

**A (Respuesta):**

    - Deportes y Electrónica lideran las categorías
    - Las regiones tienen desempeño relativamente balanceado
    - La estacionalidad representa el mayor factor de variabilidad

Las acciones sugeridas incluyen:

    - Diseñar estrategias comerciales específicas para invierno
    - Impulsar promociones en categorías con mayor potencial
